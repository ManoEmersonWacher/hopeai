```python
# import the pandas library which contains all functions related to database
import pandas as pd
```

```python
# Read the data file and create a dataset
dataset = pd.read_csv('Salary_Data.csv')
dataset
```

```python
# create the input (independent) and output (dependent) variables
independent = dataset[['YearsExperience']]
independent
```

```python
dependent = dataset[['Salary']]
dependent
```

```python
# create the training dataset and test dataset
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(independent, dependent, test_size=0.30, random_state=0)
```

```python
# select the LinearRegression model from sklearn library
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor
```

```python
# train the  model using the training dataset
regressor.fit(X_train, y_train)
```

```python
#Now we have the weight and bias calculated and available
weight = regressor.coef_
weight
```

```python
bias = regressor.intercept_
bias
```

```python
#Since we have weight and bias, we can now do the prediction
y_predict = regressor.predict(X_test)
```

```python
#We did the prediction, let's do the evaluation of the model
from sklearn.metrics import r2_score
r_score = r2_score(y_test, y_predict)
r_score
```

```python
#our model predicts nearly 97% which shows its a good model. Lets save the model
import pickle
```

```python
#filename of the linear regression model
filename = "Finalized_Linear_Regression_Model.sav"
```

```python
#save the model
pickle.dump(regressor, open(filename, 'wb'))
```

```python
# load the saved model into an object
loaded_model = pickle.load(open(filename, 'rb'))
```

```python
# From the loaded model predict the salary based on the year of experience
result = loaded_model.predict([[10]])
```
